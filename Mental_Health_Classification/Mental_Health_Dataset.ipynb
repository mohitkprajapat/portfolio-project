<h1 style="color: #2F4F4F;">Emotion Analysis for Mental Health Issues: Model Overview</h1>

<p>This model is based on the <strong><a href="https://www.kaggle.com/datasets/suchintikasarkar/sentiment-analysis-for-mental-health" target="_blank">Sentiment Analysis for Mental Health Issues dataset</a></strong> from Kaggle. The dataset contains text data related to mental health issues, and our goal is to classify these texts into categories such as Anxiety, Bipolar, Depression, Stress, and others using advanced machine learning techniques.</p>

<h3 style="color: #4682B4;">Objective of the Model</h3>
<p>The goal of the model is to classify mental health conditions based on text data using two main components:
   <ol>
     <li><strong>BERT (Bidirectional Encoder Representations from Transformers)</strong> for feature extraction from text.</li>
     <li><strong>Neural Network (NN)</strong> for prediction based on the features extracted by BERT.</li>
   </ol>
</p>

<h3 style="color: #4682B4;">How Does the Model Work?</h3>

<p>The model consists of two major components:</p>
<ol>
  <li><strong>BERT Embeddings for Feature Extraction:</strong>
      <ul>
        <li>BERT is a pre-trained transformer model that excels in understanding the context of words in a sentence by looking at the words before and after them (bidirectional). This makes BERT particularly suitable for tasks like sentiment analysis, where the context of words is essential.</li>
        <li>In this project, BERT is used to transform raw text data into a fixed-length vector representation, called embeddings. These embeddings capture the semantic meaning of the text and serve as features for the next step of the model.</li>
        <li>We use the pre-trained BERT model (`bert-base-uncased`) to get these embeddings, which are then passed to the neural network for further processing.</li>
      </ul>
  </li>

  <li><strong>Neural Network for Prediction:</strong>
      <ul>
        <li>After BERT generates the embeddings for each piece of text, a neural network is used to predict the corresponding mental health condition. The neural network is a simple feed-forward architecture with a Dense layer followed by a Softmax activation function.</li>
        <li>The network has two main layers: a Dense layer with 128 units and ReLU activation, and another Dense layer with 7 output units (one for each category) and Softmax activation to predict the probabilities of each class.</li>
        <li>Training is done using the cross-entropy loss function, which is suitable for multi-class classification tasks, and the Adam optimizer for efficient weight updates during training.</li>
      </ul>
  </li>
</ol>

<h3 style="color: #4682B4;">Why Use BERT?</h3>
<p>BERT is widely recognized for its state-of-the-art performance on a variety of NLP tasks, as it understands text context better than traditional methods. By using BERT embeddings, we ensure that the model captures nuanced details from the text, which are crucial in understanding emotional and mental health states expressed in user-generated text.</p>

<h3 style="color: #4682B4;">Why Use a Neural Network?</h3>
<p>The neural network is used to learn from the embeddings generated by BERT and classify the data into one of the mental health categories. The network is trained on the features extracted from BERT and learns to make predictions based on the patterns in the data. The use of a neural network allows the model to handle the high-dimensional embeddings efficiently and make accurate predictions.</p>

<h3 style="color: #4682B4;">Conclusion</h3>
<p>The combination of BERT for text understanding and a neural network for classification creates a powerful model for predicting mental health conditions from text. BERT’s embeddings provide rich, contextual features that help the neural network make informed predictions, making this approach highly effective for sentiment and emotion analysis.</p>


```python
import pandas as pd
import numpy as np
import math

from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix

import torch
from torch.utils.data import DataLoader, Dataset
from transformers import BertModel, BertTokenizer

import seaborn as sns
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

<h3 style="color: #4169E1;"> Data Loading & Preprocessing</h3>

<p>
We start by loading the dataset <code>Combined Data.csv</code> which contains mental health-related user statements and their corresponding emotional status labels.
</p>

<ul>
  <li><code>df = pd.read_csv("Combined Data.csv")</code>: Reads the CSV file into a pandas DataFrame.</li>
  <li><code>x = list(df.statement)</code>: Extracts the user statements as input features for the model.</li>
  <li><code>y = df.status.values.reshape(-1,1)</code>: Extracts the emotion labels and reshapes them into a column vector.</li>
</ul>

<p>
Since the labels are categorical (e.g., 'joy', 'anger', 'sadness'), we apply <strong>One-Hot Encoding</strong>:
</p>

<ul>
  <li><code>OneHotEncoder(sparse_output=False)</code>: Converts each label into a one-hot encoded vector.</li>
  <li><code>enc.fit_transform(y)</code>: Transforms the labels into a format suitable for training a classification model.</li>
  <li><code>enc.categories_</code>: Displays the list of unique emotion categories found in the data.</li>
</ul>

<p>
Finally, we split the dataset into training and testing sets using <code>train_test_split</code>, reserving 20% of the data for evaluation.
</p>


```python
df = pd.read_csv("Combined Data.csv")
```

```python
enc = OneHotEncoder(sparse_output=False)
```

```python
x = list(df.statement)
y = df.status.values.reshape(-1,1)
```

```python
y_enc = enc.fit_transform(y)
```

```python
cats = enc.categories_[0]
print(cats)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y_enc,test_size = 0.2)
```

<h3 style="color: #4169E1;">BERT Tokenization and Embedding Extraction</h3>

<p>
In this section, we use a pre-trained <strong>BERT model</strong> to convert user statements into meaningful numerical representations (embeddings), which capture the semantic and emotional context of the text.
</p>

<ul>
  <li><code>BertTokenizer.from_pretrained('bert-base-uncased')</code>: Loads a tokenizer to split and format input text for BERT.</li>
  <li><code>BertModel.from_pretrained('bert-base-uncased')</code>: Loads the pre-trained BERT model to extract contextual embeddings.</li>
</ul>

<h4 style="color: #4682B4;">Custom Dataset & Dataloader</h4>

<p>
We create a custom <code>TextDataset</code> class to wrap our training and test data for PyTorch's DataLoader:
</p>

<ul>
  <li>Handles both text and numerical features.</li>
  <li>Returns input statements and their corresponding one-hot encoded emotion labels as PyTorch tensors.</li>
</ul>

<p>
We then wrap the training and test data into <code>DataLoader</code> objects to efficiently batch and iterate through the data.
</p>

<h4 style="color: #4682B4;">Feature Extraction using BERT</h4>

<p>
Using the BERT model in evaluation mode (<code>bert.eval()</code>), we pass batches of text through the model to extract <strong>pooled embeddings</strong>, which represent the entire sentence.
</p>

<ul>
  <li><code>tokenizer(..., return_tensors='pt')</code>: Converts the input text into token IDs and attention masks for BERT.</li>
  <li><code>bert(...)</code>: Processes the tokenized inputs to generate hidden states and pooled output.</li>
  <li><code>outputs.pooler_output</code>: A fixed-size embedding vector for each sentence (used for classification).</li>
  <li>We store these embeddings for both training and test sets using <code>torch.cat</code> and convert them to NumPy arrays for further use.</li>
</ul>

<p>
These embeddings will serve as input features for downstream classification models to predict the emotion labels.
</p>


```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert = BertModel.from_pretrained('bert-base-uncased')
```

```python
class TextDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels
        self.istext = isinstance(features[0],str)

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {}
        if self.istext:
            item['features'] = self.features[idx]
        else:
            item['features'] = torch.tensor(self.features[idx])
        item['label'] = torch.tensor(self.labels[idx])
        return item

dataset_train = TextDataset(x_train, y_train)
dataloader_train = DataLoader(dataset_train, batch_size=32)
dataset_test = TextDataset(x_test, y_test)
dataloader_test = DataLoader(dataset_test, batch_size=32)
```

```python
all_embeddings = []
all_labels = []

bert.eval()
with torch.no_grad():
    for batch in dataloader_train:
        encodings = tokenizer(batch["features"], padding=True, truncation=True, return_tensors='pt')
        outputs = bert(input_ids=encodings['input_ids'], attention_mask=encodings['attention_mask'])
        pooled_output = outputs.pooler_output
        all_embeddings.append(pooled_output)
        all_labels.append(batch['label'])

all_embeddings_train = torch.cat(all_embeddings, dim=0).numpy()
all_labels_train = torch.cat(all_labels, dim=0).numpy()
```

```python
all_embeddings = []
all_labels = []

bert.eval()
with torch.no_grad():
    for batch in dataloader_test:
        encodings = tokenizer(batch["features"], padding=True, truncation=True, return_tensors='pt')
        outputs = bert(input_ids=encodings['input_ids'], attention_mask=encodings['attention_mask'])
        pooled_output = outputs.pooler_output
        all_embeddings.append(pooled_output)
        all_labels.append(batch['label'])

all_embeddings_test = torch.cat(all_embeddings, dim=0).numpy()
all_labels_test = torch.cat(all_labels, dim=0).numpy()
```

<h3 style="color: #4169E1;">Saving and Loading BERT Embeddings</h3>

<p>
After extracting the BERT-based sentence embeddings and emotion labels for both training and test sets, we save them to disk for future use. This avoids repeating the expensive embedding process every time the notebook runs.
</p>

<ul>
  <li><code>np.save()</code>: Saves the NumPy arrays (features and labels) to <code>.npy</code> files on disk.</li>
  <li><code>features_file</code>, <code>targets_file</code>, etc. are filenames for storing training and test data separately.</li>
</ul>

<p>
Later, we can load these saved arrays using <code>np.load()</code> to directly access the embeddings and targets without re-running the BERT model.
</p>

<ul>
  <li><code>np.load(features_file)</code>: Loads the saved training features (embeddings).</li>
  <li><code>np.load(targets_file)</code>: Loads the corresponding one-hot encoded training labels.</li>
  <li>Similar operations are performed for the test set.</li>
</ul>

<p>
This practice improves efficiency, especially when experimenting with multiple classification models using the same embeddings.
</p>


```python
features_file = r'loaded_features.npy'
targets_file = r'loaded_targets.npy'
test_features_file = r'loaded_test_features.npy'
test_targets_file = r'loaded_test_targets.npy'

np.save(features_file,all_embeddings_train)
np.save(targets_file,all_labels_train)
np.save(test_features_file,all_embeddings_test)
np.save(test_targets_file,all_labels_test)
```

```python
loaded_training_features = np.load(features_file)
loaded_training_targets = np.load(targets_file)
loaded_test_features = np.load(test_features_file)
loaded_test_targets = np.load(test_targets_file)
```

<h3 style="color: #4169E1;">Model Building and Training</h3>

<p>
In this section, we define and train a <strong>neural network model</strong> for emotion classification using the pre-trained BERT embeddings as input features.
</p>

<h4 style="color: #4682B4;">Model Architecture</h4>

<ul>
  <li><strong>Input Layer:</strong> The model takes 768-dimensional BERT embeddings as input (shape <code>(768,)</code>).</li>
  <li><strong>Hidden Layer:</strong> A <code>Dense</code> layer with 128 units and ReLU activation for learning non-linear relationships.</li>
  <li><strong>Dropout:</strong> A dropout layer with 0.3 probability to prevent overfitting during training.</li>
  <li><strong>Output Layer:</strong> A final <code>Dense</code> layer with 7 units (one for each emotion) and softmax activation to output a probability distribution across the 7 emotion categories.</li>
</ul>

<h4 style="color: #4682B4;">Model Compilation</h4>

<p>
We compile the model with the following settings:
</p>

<ul>
  <li><strong>Optimizer:</strong> Adam optimizer for efficient gradient descent.</li>
  <li><strong>Loss Function:</strong> Categorical cross-entropy, since we have multiple classes (7 emotions).</li>
  <li><strong>Metrics:</strong> Accuracy, to monitor model performance during training.</li>
</ul>

<h4 style="color: #4682B4;">Model Training</h4>

<p>
The model is trained on the preprocessed BERT embeddings with the following settings:
</p>

<ul>
  <li><strong>Batch Size:</strong> 32, the number of samples processed before updating the model's weights.</li>
  <li><strong>Validation Split:</strong> 20% of the training data is used for validation during training.</li>
  <li><strong>Epochs:</strong> A maximum of 1000 epochs, with early stopping after 50 epochs of no improvement in validation accuracy.</li>
  <li><strong>Callbacks:</strong> 
    <ul>
      <li><code>EarlyStopping</code>: Stops training early if the validation accuracy doesn't improve for 50 epochs, restoring the best weights.</li>
      <li><code>ModelCheckpoint</code>: Saves the model weights at regular intervals (every 100 epochs) during training.</li>
    </ul>
  </li>
</ul>

<p>
Once training is complete, we save the trained model to <code>my_model_short.keras</code> for later use or deployment.
</p>


```python
model = Sequential([
    Input(shape=(768,)),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(7, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
batch_size = 32
val_split = 0.2
num_batch = math.ceil((loaded_training_features.shape[0] * (1-val_split)) / batch_size)

epoch_save_freq = 100
batch_save_freq = num_batch * epoch_save_freq

ES = EarlyStopping(monitor='val_accuracy', patience=50, restore_best_weights=True)
CP = ModelCheckpoint(r'model_short_epoch_{epoch:02d}.keras',save_weights_only=False,save_best_only=False,save_freq=batch_save_freq)
```

```python
history_short = model.fit(loaded_training_features, loaded_training_targets, 
                    epochs=1000, 
                    batch_size=batch_size, 
                    validation_split=val_split,
                    verbose = 1,
                    callbacks = [ES,CP])
model.save(r'my_model_short.keras')
```

<h3 style="color: #4169E1;">Model Evaluation and Confusion Matrix</h3>

<p>
Once the model is trained, we evaluate its performance on the test set by making predictions and comparing them with the true labels.
</p>

<h4 style="color: #4682B4;">Predictions and Accuracy</h4>

<ul>
  <li><code>model.predict(loaded_test_features)</code>: Uses the trained model to predict emotion classes for the test set.</li>
  <li><code>y_pred.argmax(axis=1)</code>: Converts the predicted probabilities into class labels by selecting the class with the highest probability.</li>
  <li><code>y_true = loaded_test_targets.argmax(axis=1)</code>: Converts the one-hot encoded true labels into class labels.</li>
  <li><strong>Accuracy:</strong> We calculate the model's accuracy by comparing the predicted labels with the true labels using <code>accuracy_score</code>.</li>
</ul>

<p>
The accuracy score is printed to show how well the model performed on the test set.
</p>

<h4 style="color: #4682B4;">Confusion Matrix</h4>

<p>
To better understand the performance, we visualize a <strong>confusion matrix</strong>, which shows how many times each class was correctly or incorrectly predicted.
</p>

<ul>
  <li><code>confusion_matrix(y_true, y_predicted, normalize = 'true')*100</code>: Computes the confusion matrix comparing the true labels and predicted labels.</li>
  <li><strong>Heatmap:</strong> We use Seaborn's <code>heatmap()</code> function to visualize the confusion matrix as a percentage. The axes represent the actual and predicted emotion classes.</li>
</ul>

<p>
The heatmap provides an intuitive view of where the model is making correct and incorrect predictions across all classes.
</p>


```python
y_pred = model.predict(loaded_test_features)
```

```python
y_predicted = y_pred.argmax(axis = 1)
y_true = loaded_test_targets.argmax(axis = 1)
```

```python
acc = accuracy_score(y_true, y_predicted)
print(f"Test Accuracy: {acc:.4f}")
```

```python
con_mat = confusion_matrix(y_true, y_predicted, normalize = 'true')*100
sns.heatmap(con_mat, annot = True, fmt = "0.1f", xticklabels = cats, yticklabels = cats)
plt.xlabel("Predictions")
plt.ylabel("Actual")
plt.title("Percentage of Each Actual Class Predicted as All Classes")
plt.show()
```
