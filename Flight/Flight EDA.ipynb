<h2>Handling Missing Values and Data Cleaning</h2>

<p>To identify missing values in the dataset, I use <code>df.isnull().sum()</code>, which returns the count of null values in each column.</p>

<p>For the column <code>CRS_ELAPSED_TIME</code>, I handle missing values by calculating the difference between the scheduled arrival (<code>CRS_ARR_TIME</code>) and departure times (<code>CRS_DEP_TIME</code>), converting the time difference into minutes. This ensures that missing values in this column are filled with a reasonable estimate.</p>

<p>Next, I check for missing values again, but this time filtering out diverted and canceled flights using <code>df[(df.DIVERTED == 0) & (df.CANCELLED == 0)].isnull().sum()</code>. This helps assess missing data only for flights that were actually operated.</p>

<p>To clean the dataset, I remove all diverted and canceled flights using <code>df.drop(df[df.DIVERTED == 1].index, inplace=True)</code> and <code>df.drop(df[df.CANCELLED == 1].index, inplace=True)</code>. Additionally, I drop the <code>'CANCELLATION_CODE'</code> column since it's no longer needed after removing canceled flights.</p>

<p>Finally, I check for missing values specifically in columns related to delay analysis using <code>df[df.DELAY_DUE_LATE_AIRCRAFT.notnull()][col_delay].isnull().sum()</code>. This helps in understanding if delay-related features have missing values and need further processing.</p>

<h2>Airline-Wise Delay Analysis</h2>

<p>To analyze airline delays, I first compute the total number of delayed flights for each airline using <code>df_delayed.AIRLINE.value_counts().reset_index()</code>. Similarly, I calculate the total number of flights per airline using <code>df.AIRLINE.value_counts().reset_index()</code>.</p>

<p>Next, I merge these two dataframes to get a consolidated view of delayed and total flights per airline using <code>df_airline_delayed.merge(df_airline_total, on='AIRLINE', suffixes=('_Delayed', '_Total'))</code>.</p>

<p>To understand how long flights are typically delayed, I calculate the average arrival delay per airline using <code>df_delayed.groupby('AIRLINE')['ARR_DELAY'].mean().reset_index().sort_values(by='ARR_DELAY')</code> and merge this information into the main dataframe.</p>

<h2>Visualization of Airline Delays</h2>

<p>To better understand the airline-wise delay patterns, I create four bar charts using Seaborn:</p>

<ul>
    <li><strong>Top-left chart:</strong> Shows the percentage of flights delayed for the top 15 airlines with the highest delay percentage.</li>
    <li><strong>Top-right chart:</strong> Displays the mean delay time of delayed flights for these airlines.</li>
    <li><strong>Bottom-left chart:</strong> Represents the total number of flights operated by each airline in the top 15.</li>
    <li><strong>Bottom-right chart:</strong> Highlights the number of flights that were delayed for each airline in the top 15.</li>
</ul>

<p>The charts provide insights into which airlines experience the highest proportion of delays, their average delay time, and how their total flight volume relates to the number of delays.</p>

<p><code>plt.tight_layout(pad=2.0)</code> ensures that all subplots are well-spaced, and <code>plt.show()</code> displays the final visualizations.</p>


```python
df_airline_delayed = df_delayed.AIRLINE.value_counts().reset_index()
df_airline_total = df.AIRLINE.value_counts().reset_index()

df_airline = df_airline_delayed.merge(df_airline_total,on = 'AIRLINE', suffixes = ('_Delayed','_Total'))

df_airline_delay_time = df_delayed.groupby('AIRLINE')['ARR_DELAY'].mean().reset_index().sort_values(by = 'ARR_DELAY')

df_airline = df_airline.merge(df_airline_delay_time,on = 'AIRLINE')

df_airline['Percentage'] = (df_airline.count_Delayed / df_airline.count_Total)*100
df_airline_top = df_airline.sort_values(by = 'Percentage', ascending = False).head(15)

plt.figure(figsize=(16, 16))

plt.subplot(2,2,1)
sns.barplot(data = df_airline_top, y = 'AIRLINE', x = 'Percentage', color=custom_palette[0], orient='h')
plt.xlabel("Percentage of flights delayed")
plt.ylabel("Airline")
plt.grid(visible=True, which='major', axis='x')
plt.title("Percent of delayed flights for \nTop 15 Airline with highest percentage flights delayed at arrival")

plt.subplot(2,2,2)
sns.barplot(data = df_airline_top, y = 'AIRLINE', x = 'ARR_DELAY', color=custom_palette[3], orient='h')
plt.xlabel("Mean Delay")
plt.ylabel("Airline")
plt.grid(visible=True, which='major', axis='x')
plt.title("Average delay time of delayed flights for \nTop 15 Airline with highest percentage flights delayed at arrival")

plt.subplot(2,2,3)
sns.barplot(data = df_airline_top, y = 'AIRLINE', x = 'count_Total', color=custom_palette[1], orient='h')
plt.xlabel("Number of Flights")
plt.ylabel("Airline")
plt.grid(visible=True, which='major', axis='x')
plt.title("Total number of flights for \nTop 15 Airline with highest percentage flights delayed at arrival")

plt.subplot(2,2,4)
sns.barplot(data = df_airline_top, y = 'AIRLINE', x = 'count_Delayed', color=custom_palette[2], orient='h')
plt.xlabel("Number of Flights")
plt.ylabel("Airline")
plt.grid(visible=True, which='major', axis='x')
plt.title("Number of delayed flights for \nTop 15 Airline with highest percentage flights delayed at arrival")

plt.tight_layout(pad=2.0)
plt.show()
```

<p style="background-color:LightGray; color:black">
Allegiant Air has the highest percentage of delayed flights, while JetBlue Airways and Hawaiian Airlines experience the longest average delays (~50 minutes). <h2>Route-Wise Delay Analysis</h2>

<p>To analyze delays based on flight routes, I first compute the total number of delayed flights for each route using <code>df_delayed.ROUTE.value_counts().reset_index()</code>. <p>To understand how long flights are typically delayed on different routes, I calculate the average arrival delay per route using <code>df_delayed.groupby('ROUTE')['ARR_DELAY'].mean().reset_index().sort_values(by='ARR_DELAY')</code> and merge this information into the main dataframe.</p>

<p>I then compute the percentage of delayed flights for each route using <code>(df_route.count_Delayed / df_route.count_Total) * 100</code> and extract the top 20 busiest routes based on total flight volume.</p>

<h2>Visualization of Route Delays</h2>

<p>To better understand the delay trends across routes, I create four bar charts using Seaborn:</p>

<ul>
    <li><strong>Top-left chart:</strong> Shows the total number of flights for the 20 busiest routes.</li>
    <li><strong>Top-right chart:</strong> Displays the number of delayed flights for these routes.</li>
    <li><strong>Bottom-left chart:</strong> Represents the percentage of delayed flights per route.</li>
    <li><strong>Bottom-right chart:</strong> Highlights the average delay time of delayed flights per route.</li>

```python
df_route_delayed = df_delayed.ROUTE.value_counts().reset_index()
df_route_total = df.ROUTE.value_counts().reset_index()

df_route = df_route_delayed.merge(df_route_total,on = 'ROUTE', suffixes = ('_Delayed','_Total'))

df_route_delay_time = df_delayed.groupby('ROUTE')['ARR_DELAY'].mean().reset_index().sort_values(by = 'ARR_DELAY')

df_route = df_route.merge(df_route_delay_time,on = 'ROUTE')

df_route['Percentage'] = (df_route.count_Delayed / df_route.count_Total)*100
df_route_top = df_route.sort_values(by = 'count_Total', ascending = False).head(20)

plt.figure(figsize=(16, 16))

plt.subplot(2,2,1)
sns.barplot(data = df_route_top, y = 'ROUTE', x = 'count_Total', color=custom_palette[1], orient='h')
plt.xlabel("Number of Flights")
plt.ylabel("Route")
plt.grid(visible=True, which='major', ls = ':', axis='x')
plt.title("Total number of flights for \nTop 20 Busiest Route")

plt.subplot(2,2,2)
sns.barplot(data = df_route_top, y = 'ROUTE', x = 'count_Delayed', color=custom_palette[2], orient='h')
plt.xlabel("Number of Flights")
plt.ylabel("Route")
plt.grid(visible=True, which='major', ls = ':', axis='x')
plt.title("Number of delayed flights for \nTop 20 Busiest Route")

plt.subplot(2,2,3)
sns.barplot(data = df_route_top, y = 'ROUTE', x = 'Percentage', color=custom_palette[0], orient='h')
plt.xlabel("Percentage of flights delayed")
plt.ylabel("Route")
plt.grid(visible=True, which='major', ls = ':', axis='x')
plt.title("Percent of delayed flights for \nTop 20 Busiest Route")

plt.subplot(2,2,4)
sns.barplot(data = df_route_top, y = 'ROUTE', x = 'ARR_DELAY', color=custom_palette[3], orient='h')
plt.xlabel("Mean Delay")
plt.ylabel("Route")
plt.grid(visible=True, which='major', ls = ':', axis='x')
plt.title("Average delay time of delayed flights for \nTop 20 Busiest Route")


plt.tight_layout(pad=2.0)
plt.show()
```

<h2>Flight Delay Trends Over Time</h2>

<p>To analyze trends in flight delays over time, I calculate the average departure and arrival delays per day and visualize them using a time series plot.</p>

<h3>Data Processing</h3>

<ul>
    <li>Using <code>groupby("FL_DATE")</code>, I compute the daily mean of departure (<code>DEP_DELAY</code>) and arrival (<code>ARR_DELAY</code>) delays for delayed flights.</li>
    <li>To smooth out daily fluctuations, I apply a 7-day rolling average with <code>rolling(window=7, min_periods=1).mean()</code>.</li>
</ul>

<h3>Visualization</h3>

<ul>
    <li>The smoothed departure delay trend is plotted in <span style="color: custom_palette[1];">one color</span>.</li>
    <li>The smoothed arrival delay trend is plotted in <span style="color: custom_palette[4];">another color</span>.</li>
    <li>Both plots help identify patterns such as seasonal trends or sudden spikes in delays.</li>
</ul>

<h3>Key Features</h3>

<ul>
    <li><code>plt.plot()</code> is used to plot the delay trends over time.</li>
    <li>Grid lines are added to improve readability.</li>
    <li><code>plt.xticks(rotation=45)</code> ensures date labels are readable.</li>
    <li>The title and axis labels are included for clarity.</li>
</ul>

<p>These insights can help in identifying periods of higher delays, possibly due to weather conditions, peak travel seasons, or operational inefficiencies.</p>


```python
df_daily_delay = df_delayed.groupby("FL_DATE")[["DEP_DELAY", "ARR_DELAY"]].mean().reset_index()

df_daily_delay["DEP_DELAY_SMOOTH"] = df_daily_delay["DEP_DELAY"].rolling(window=7, min_periods=1).mean()
df_daily_delay["ARR_DELAY_SMOOTH"] = df_daily_delay["ARR_DELAY"].rolling(window=7, min_periods=1).mean()
```

```python
plt.figure(figsize=(16,6))

plt.plot(df_daily_delay["FL_DATE"], df_daily_delay["DEP_DELAY_SMOOTH"], label="Departure Delay", color=custom_palette[1])
plt.plot(df_daily_delay["FL_DATE"], df_daily_delay["ARR_DELAY_SMOOTH"], label="Arrival Delay", color=custom_palette[4])

plt.xlabel("Date")
plt.ylabel("Average Delay (minutes)")
plt.title("Flight Delay Trends Over Time For Delayed Flights")
plt.legend()
plt.grid(visible=True, which='major', ls = ':', axis='both')
plt.xticks(rotation=45)
plt.show()
```

<p style="background-color:LightGray; color:black">The trend in flight delays shows seasonal fluctuations, with noticeable peaks in delay times occurring periodically. There was a significant drop in delays around early 2021, possibly due to reduced air traffic from pandemic-related restrictions. However, delays increased again post-2021, with spikes in early 2023, indicating potential operational disruptions. Arrival delays (green) tend to be slightly higher than departure delays (orange), likely due to cascading effects from previous flight disruptions.</p>
